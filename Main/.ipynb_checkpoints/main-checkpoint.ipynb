```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# temporary file locations (DELETE BEFORE SUBMISSION)
mouse_metadata_path = '/Users/Brianna/Desktop/JHUmaterials/Submissions/matplotlib-challenges/Resources/Mouse_metadata.csv'
study_results_path = '/Users/Brianna/Desktop/JHUmaterials/Submissions/matplotlib-challenges/Resources/Study_results.csv'
# Study data files (UNCOMMENT BEFORE SUBMISSION)
# mouse_metadata_path = "Resources/Mouse_metadata.csv"
# study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_df = pd.merge(mouse_metadata, study_results, on='Mouse ID', how='inner')
# Display the data table for preview
mouse_df
```

```python
# Checking the number of mice.
num_of_mice = len(mouse_df['Mouse ID'])
num_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicates_df = mouse_df[mouse_df.duplicated(subset=['Mouse ID','Timepoint'], keep=False)]
duplicates_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
total_duplicates_df = mouse_df[mouse_df.duplicated(subset=['Mouse ID'])]
total_duplicates_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = mouse_df.drop_duplicates(subset=['Mouse ID'])
clean_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
unique_mice_clean = len(clean_mouse_df['Mouse ID'])
unique_mice_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
    # mean, median, variance, standard deviation, and SEM of the tumor volume. 
    # Assemble the resulting series into a single summary dataframe.

mean = mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
st_dev = mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
SEM = mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summary_stats = pd.DataFrame({'Mean': mean, 
                              'Median': median, 
                              'Variance': variance, 
                              'Standard Deviation': st_dev, 
                              'SEM': SEM})
summary_stats
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.

# group mouse data by drug regimen
grouped_drugs = pd.DataFrame(mouse_df.groupby(['Drug Regimen']).count()).reset_index()

# grab drug regimen and mouse id columns
regimen_df = grouped_drugs[['Drug Regimen', 'Mouse ID']]
regimen_df = regimen_df.set_index('Drug Regimen')

# create bar chart
regimen_df.plot(kind="bar", figsize=(8,3), color='green')
plt.title('Measurement Count Based on Drug Treatment')
plt.show()
```

```python
regimen_df
```

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pyplot.

# Set x axis and tick locations
x_axis = np.arange(len(regimen_df))
tick_locations = [value+0.4 for value in x_axis]
drug = [i for i in regimen_df]

# Create a list indicating where to write x labels and set figure size to adjust for space

plt.figure(figsize=(8,3))
plt.bar(x_axis, regimen_df['Mouse ID'], color='g')
plt.xticks(tick_locations, drug, rotation="vertical")

# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(regimen_df["Mouse ID"])+10)

# Set a Title and labels
plt.title('Measurement Count Based on Drug Treatment')
plt.xlabel('Drug Regimen')
plt.ylabel('Measurement Count')

# Save our graph and show the grap
#plt.tight_layout()
#plt.savefig("../Images/avg_state_rain.png")
plt.show()
```
